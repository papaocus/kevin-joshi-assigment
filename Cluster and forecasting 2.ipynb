```python
import pandas as pd

# Replace "path/to/file.xlsx" with the actual path to your file
df = pd.read_excel("D:\Kevin Joshi\cons_data_all_res(1).xlsx")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt



for i in range(1, 13):
    plt.plot(df['Time'], df[f"r{i}"], label=f"Rainfall {i}")

plt.plot(df['Time'], df['temp'], label="Temperature")


plt.xlabel('Time')
plt.ylabel('Temperature/Rainfall')

plt.legend()


plt.show()
```

```python
from sklearn.cluster import KMeans
rainfall = df.loc[:, 'r1':'r12']
kmeans = KMeans(n_clusters=5)
kmeans.fit(rainfall)
labels = kmeans.labels_
df['cluster'] = labels
print(df['cluster'].value_counts())
```
