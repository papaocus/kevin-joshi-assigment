```python
##Label encoding the catogoerical coloumns
from sklearn.preprocessing import LabelEncoder




# Create a LabelEncoder object
le = LabelEncoder()

# Fit the LabelEncoder object to the column to be encoded
df["month_name"] = le.fit_transform(df["month_name"])
df["busday"] = le.fit_transform(df["busday"])
df["season"] = le.fit_transform(df["season"])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['r1'])
plt.xlabel('Time')
plt.ylabel('r1')
plt.title('Time Series Data')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index, df['temp'], label='Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
df.boxplot(column=['temp'], by='month_name', ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
plt.show()

```

```python
plt.figure(figsize=(10,6))
plt.hist(df['temp'], bins=30)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['temp'], model='additive', period=365)
result.plot()
plt.show()
```

```python
from pandas.plotting import lag_plot

fig, ax = plt.subplots(figsize=(10,6))
lag_plot(df['temp'], ax=ax)
ax.set_xlabel('Temperature(t)')
ax.set_ylabel('Temperature(t+1)')
plt.show()

```

```python
fig, ax = plt.subplots(figsize=(10,6))
df.boxplot(column=['temp'], by='month_name', ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df[['temp', 'humid', 'r1', 'r2', 'r3']], figsize=(10, 10))
plt.show()
```

```python
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
window = 30
rolling_mean = df['temp'].rolling(window=window).mean()

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['temp'], label='Temperature')
ax.plot(rolling_mean, label=f'Rolling Mean ({window} days)')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
ax.legend()
plt.show()
```

```python
diff = df['temp'].diff()
diff = diff.dropna()

fig, ax = plt.subplots(figsize=(10,6))
diff.plot(kind='hist', ax=ax)
ax.set_xlabel('Temperature Difference')
ax.set_ylabel('Frequency')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
df.loc[:, 'r1':'r12'].plot(kind='hist', alpha=0.5, ax=ax)
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.legend(['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10', 'r11', 'r12'])
plt.show()
```

```python
corr_matrix = df.loc[:, 'r1':'r12'].corr()

fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
window = 30
rolling_means = df.loc[:, 'r1':'r12'].rolling(window=window).mean()

fig, ax = plt.subplots(figsize=(10, 6))
df.loc[:, 'r1':'r12'].plot(ax=ax)
rolling_means.plot(ax=ax)
ax.set_xlabel('Date')
ax.set_ylabel('Value')
ax.legend(['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10', 'r11', 'r12',
            f'r1-r12 Rolling Mean ({window} days)'])
plt.show()
```
