```python
##A rand select one from r1 to r12
##Apply prediction for the next hour
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_excel("D:\Kevin Joshi\cons_data_all_res(1).xlsx")
```

```python
df.head()
```

```python
rand_r = 'r' + str(np.random.randint(1, 13))
time_series = pd.Series(df[rand_r].values, index=pd.to_datetime(df['Time']))

model = sm.tsa.ARIMA(time_series, order=(1, 0, 1))
results = model.fit()


next_hour = time_series.index[-1] + pd.Timedelta(hours=1)
forecast = results.forecast(steps=1)

print("The forecasted value for the next hour for", rand_r, "is:", forecast[0])
```
